# This is our O(n^2) function.

### We are going to call it many times to create plottable data points.


```python
def log_pairs_of_lists(list_to_log: list[str]) -> dict[str, int]:
    operations_counter = 0

    for i in range(len(list_to_log)):
        operations_counter += 1 # for every operation we increase the counter
        for j in range(len(list_to_log)):
            operations_counter += 1 # for every operation we increase the counter
            # use line below for debugging purposes
            # print(f"{list_to_log[i]}, {list_to_log[j]}") 

    dictionary = dict(no_items = len(list_to_log), operations_counter = operations_counter)
    return dictionary
```

# Setup

### Create Functions that Generate plottable data

First lets create a function that fills a list of random length, with strings.

```python
def fill_list(name: str, number_of_items: int) -> list[str]:
    output = []

    for i in range(number_of_items):
        output.append(name)
        
    return output
```

Then create a function that creates 10 of those lists

```python
import random

def create_100_random_lists() -> list[list[str]]:
    random_lists = []

    for i in range(100):
        random_lists.append(fill_list(
            name = "string", 
            number_of_items = random.randint(10,100)))            
    
    return random_lists
```

Finally we want to create a function that returns to us the plottable results.

```python
def get_plottable_result(random_lists: list[list[str]]) -> list[dict[str, int]]:
    plottable = []
    for random_list in random_lists:
        dictionary = log_pairs_of_lists(random_list)
        plottable.append(dictionary)
        
    return plottable
```

### Get the plottable data

```python
plottable = get_plottable_result(create_100_random_lists())
```

### Create plot function 

```python
import matplotlib.pyplot as plt

def plot_results(plottable: list[dict[str, int]]) -> None:
    x_values = [data_point["no_items"] for data_point in plottable]
    y_values = [data_point["operations_counter"] for data_point in plottable]

    plt.plot(x_values, y_values, 'o')
    plt.xlabel('Number of Items')
    plt.ylabel('Operations')
    plt.title('Number of Items vs Operations O(n^2)')
    plt.show()
```

# Plot!

```python
plot_results(plottable)
```

Now, compare it to the O(n). Massive difference!

As it is apparent, with O(n) it took 10.000 elements in an array to get to 10.000 operations.

With O(n^2) it took 100 elements to reach the same amount of operations. 

Thats nested for-loops for ya!
