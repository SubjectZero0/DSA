# This is our O(n) function. 

### We are gonna set it up and call it many of times to get our data points

```python
def find_nemo(fish_array: list[str]) -> dict[str, int]:
    ops_counter = 0
    for fish in fish_array:
        ops_counter+=1 # for every operation we increase the counter
        
    dictionary = dict(no_items = len(fish_array), ops_counter = ops_counter)

    return dictionary
```

# Setup

*First we set up a function to fill lists with variable lengths (random)*

```python
def fill_list(name: str, number_of_items: int) -> list[str]:
    output = []

    for i in range(number_of_items):
        output.append(name)
        
    return output
```

*Then we create a function that creates 100 lists of fish.*

```python
import random

def create_100_fish_lists() -> list[list[str]]:
    fish_lists = []

    for i in range(100):
        fish_lists.append(fill_list(
            name = "nemo", 
            number_of_items = random.randint(100, 10000)))            
    
    return fish_lists
```

*Finally we want to create a function that returns to us the plottable results*

```python
def get_plottable_result(fish: list[list[str]]) -> list[dict[str, int]]:
    plottable = []
    for list_of_fish in fish:
        dictionary = find_nemo(list_of_fish)
        plottable.append(dictionary)
        
    return plottable
```

# Get the plottable results

```python
plottable = get_plottable_result(create_100_fish_lists())
```

# Plot!

```python
import matplotlib.pyplot as plt

def plot_results(plottable: list[dict[str, int]]) -> None:
    x_values = [data_point["no_items"] for data_point in plottable]
    y_values = [data_point["ops_counter"] for data_point in plottable]

    plt.plot(x_values, y_values)
    plt.xlabel('Number of Items')
    plt.ylabel('Operations')
    plt.title('Number of Items vs Operations O(n)')
    plt.show()
```

```python
plot_results(plottable)
```

### Hence we have demonstrated the linear nature of O(n)!!!
